```python
# Load the .npz file
data_imgs = np.load('kkanj-imgs.npz')
data_labels = np.load('kkanji-labels.npz')

# Access the arrays stored in the .npz files
imgs = data_imgs['arr_0']
labels = data_labels['arr_0']

# Perform operations on the loaded data
print("Shape of imgs:", imgs.shape)
print("Shape of labels:", labels.shape)



num_elements_to_print = 5
print("First", num_elements_to_print, "elements in imgs:")
print(imgs[:num_elements_to_print])


num_elements_to_print = 5
print("First", num_elements_to_print, "elements in labels:")
print(labels[:num_elements_to_print])
```

```python
from sklearn.model_selection import train_test_split


# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(imgs, labels, test_size=0.1, random_state=42)

# Printing the sizes of the resulting sets
print("Size of x_train:", len(x_train))
print("Size of y_train:", len(y_train))
print("Size of x_test:", len(x_test))
print("Size of y_test:", len(y_test))


```

```python
# normalize data from 0-255 to 0-1
x_train_og = x_train
y_train_og = y_train
y_test_og = y_test
x_test_og = x_test
x_train = x_train_og.astype(np.float32)
x_test = x_test_og.astype(np.float32)
image_size = 64*64

x_train /= 255


x_test /= 255

x_train[3500]
```

```python
from keras.utils import to_categorical # to get a one-hot incoding of the target data
num_classes = total_unique_values = len(set(labels))

y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
y_test[:4]
```

```python
type(y_train)
print(y_train.size)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_encoder = LabelEncoder()
integer_labels = label_encoder.fit_transform(y_train)

# Convert integer labels to one-hot encoding
onehot_encoder = OneHotEncoder(sparse=False, categories='auto')
onehot_labels = onehot_encoder.fit_transform(integer_labels.reshape(-1, 1))

# Print the one-hot encoded labels
print("One-hot encoded labels:")

print(onehot_labels)
```

```python
from keras.models import Sequential
# model instance
model = Sequential()
# define each layer in the network
from keras.layers import Dense # the simplest layer type

input_layer = Dense(units=512,  # 512 units (nodes) is an arbitrary number (might be ajusted later up or down to see what gives most accuracy)
                activation='sigmoid', # activation is the activation function used to pass dot product of all inputs and weights through
                input_shape=(image_size, )) #64x64 for the mnist image size (must be an iterable, therefore the comma)
```

```python
for i in (image_size,):
    print(i)
print(len((image_size,)))
```

```python
model.add(input_layer)
# add another  hidden layer
model.add(Dense(units=512,activation='sigmoid'))

#model.add(Dense(units=48,activation='relu'))

# add an output layer
model.add(Dense(units=10,
                activation='softmax', # softmax nonlinearity function for mapping the neural network activation to the categories
                #input_shape=(image_size,) # does not seem to be warranted
               ))
model.summary()
```

```python
# Now train the model
model.compile(loss='categorical_crossentropy', # loss is how to meassure how wrong the model is on its predictions
             optimizer='sgd', # "stochastic gradient descent" is a way to tell algorithm how to improve
             metrics=['accuracy'], # what do we care about in our model
             )

x_train_reshaped = x_train.reshape(-1, 64*64)
model.fit(x_train_reshaped,
         y_train,
         epochs=5,
         verbose=True,
         validation_split=0.1) # checking periodically how well we are doing
```
