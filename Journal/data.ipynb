```python
import numpy as np
import rand as rand
```

```python
import pathlib
from pathlib import Path
list(Path(".").iterdir())
from PIL import Image, ImageOps
```

```python
import tarfile
tar = tarfile.open("kkanji.tar") # Open for reading with gzip compression.
tar.extractall()
tar.close
```

```python

imgs = np.zeros((64,64), np.uint8).reshape(1,64,64) #Dummy black image to setup the ndarray
labels = np.array(['XXX'])

for folder in Path("kkanji2").iterdir():
    label = folder.__str__().split("\\")[1]
    for kanji_path in folder.iterdir():
        image = np.array(Image.open(kanji_path)).reshape(1,64,64)
        imgs = np.concatenate([imgs,image])
        labels = np.concatenate([labels,np.array([label])])

imgs = np.delete(imgs,0,0)
labels = np.delete(labels,0,0)
```

```python
np.savez_compressed('kkanj-imgs.npz', imgs)
np.savez_compressed('kkanji-labels.npz', labels)
```

```python
# Load the .npz file
data_imgs = np.load('kkanj-imgs.npz')
data_labels = np.load('kkanji-labels.npz')

# Access the arrays stored in the .npz files
imgs = data_imgs['arr_0']
labels = data_labels['arr_0']

# Perform operations on the loaded data
print("Shape of imgs:", imgs.shape)
print("Shape of labels:", labels.shape)



num_elements_to_print = 5
print("First", num_elements_to_print, "elements in imgs:")
print(imgs[:num_elements_to_print])


num_elements_to_print = 5
print("First", num_elements_to_print, "elements in labels:")
print(labels[:num_elements_to_print])




```
