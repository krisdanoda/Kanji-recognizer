```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from modules import helper_functions as hp
import numpy as np
```

```python

data_labels = np.load('../../data/kkanji-labels.npz')
labels = data_labels['arr_0']
labels = np.array(labels)
labels = np.unique(labels)
kanji_list = []
for label in labels:
    kanji_list.append( hp.to_kanji(label))
i = 0;
# List of kanji characters

# Create an empty list to store the scraped data
data = []

# Loop through the kanji list
for kanji in kanji_list:
    label = labels [i]
    # Construct the URL for the kanji page
    url = f"https://jisho.org/search/{kanji}"
    print("scraping ", i, " out of ", len(labels))
    i = i+1
    # Send a GET request to the URL and retrieve the page content
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Extract the meanings in English sense
    meanings = soup.find(class_="meanings english sense")
    if meanings:
        definition = meanings.get_text(strip=True)
        definition = definition.split(",")
    else:
        definition = ""

    # Extract the on readings
    on_readings = soup.find(class_="on readings")
    if on_readings:
        on_readings = on_readings.find_all("a")
        on_readings = [reading.get_text(strip=True) for reading in on_readings]
    else:
        on_readings = []

    # Extract the kun readings
    kun_readings = soup.find(class_="kun readings")
    if kun_readings:
        kun_readings = kun_readings.find_all("a")
        kun_readings = [reading.get_text(strip=True) for reading in kun_readings]
    else:
        kun_readings = []

    # Append the scraped data to the list
    data.append([label, kanji, definition, on_readings, kun_readings])

# Create a Pandas DataFrame from the scraped data
df = pd.DataFrame(data, columns=["unicode","Kanji", "Definition", "On Readings", "Kun Readings"])
df['Definition'] = df['Definition'].apply(lambda x: ", ".join(x) if isinstance(x, list) else "")
df['On Readings'] = df['On Readings'].apply(lambda x: ", ".join(x) if isinstance(x, list) else "")
df['Kun Readings'] = df['Kun Readings'].apply(lambda x: ", ".join(x) if isinstance(x, list) else "")
# Save the DataFrame to a CSV file
df.to_csv("kanji_data.csv", index=False)
```
